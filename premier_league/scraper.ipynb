# Premier League 
### Tabels Standings

```python
url = 'https://www.premierleague.com/tables'
```

```python
requests = requests.get(url)
```

```python
soup = BeautifulSoup(requests.content, 'html.parser')
# soup
```

```python
table = soup.find('table')
# table
```

```python
# Locate the table
table = soup.find('table')
```

```python
 # Extract column headers
headers = []
for th in table.find_all('th'):
    headers.append(th.get_text(strip=True))
```

```python
# headers
```

```python
# Extract table rows
rows = table.find_all('tr')
rows
```

```python
data = []
for row in rows[1:]:  # Skip header row
    cols = row.find_all('td')
    cols = [col.get_text(strip=True) for col in cols]
    if cols:
        data.append(cols)
data
```

```python
# print(cols[0].replace('Previous Position', ''))

# Convert to a DataFrame
df = pd.DataFrame(data, columns=headers)
df_cleaned = df.dropna()
df_cleaned
```

```python
# Update the DataFrame
for index, row in df_cleaned.iterrows():
    if index >= 9:
        df_cleaned.at[index, 'PositionPos'] = row['PositionPos'][0:2]  # Update for index >= 9
    else:
        df_cleaned.at[index, 'PositionPos'] = row['PositionPos'][0]    # Update for index < 9
```
